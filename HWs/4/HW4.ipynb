<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
با استفاده از
<code>np.random.rand</code>
و
<code>np.random.randn</code>
دو آرایه
<code>X_poly</code>
و
<code>y_poly</code>
را میسازیم که با همدیگر ارتباط خطی ندارند.
  </font>
</p>

```python
#Seed the generator
np.random.seed(42)
# Determine the number of data points to generate
data_length = 200
#Generate random data points
X_poly = 16 * np.random.rand(data_length) - 8
#equation  y = 1.9x^2 + 3.6x + 8.3
y_poly = 1.9 * X_poly**2 + 3.6*X_poly + 8.3 + np.random.randn(data_length)
y_poly
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در سلول زیر 
<code>scatter plot</code>
داده های تولید شده را رسم کنید.
  </font>
</p>

```python
#scatter plot of X_poly and y_poly
fig_scatter = px.scatter(x = X_poly , y = y_poly)
#show the plot
fig_scatter.show()
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در ادامه برای ساخت ویژگی ها با درجات بیشتر لازم است
<code>X_poly</code>
را دو بعدی کنیم.
  </font>
</p>

```python
#shape of X_poly_2d should be (data_length , 1)
X_poly_2d = X_poly.reshape((data_length , 1))
X_poly_2d
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
حالا مثل تمرین سری پیش برای رگرسیون غیر خطی ویژگی هایی از درجه های 
<code>2</code>
و
<code>3</code>
تعریف میکنیم.
  </font>
</p>

```python
#calculate degree 2
X_squared = X_poly_2d ** 2
# calculate degree 3
X_cubed = X_poly_2d ** 3
X_squared
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
در ادامه خود مقادیر
<code>X_poly_2d</code>
و درجه 
<code>2</code>
و
<code>3</code>
آن را باهم در یک آرایه دو بعدی قرار میدهیم که ستون های آن نشان دهنده درجات وِیژگی است.
  </font>
</p>

```python
X_features = np.concatenate([X_poly_2d, X_squared, X_cubed] ,axis=1)
X_features
```

<p dir=rtl style="direction: rtl;text-align: right;line-height:200%;font-family:vazir;font-size:medium">
<font face="vazir" size=3>
برای 
<code>Overfit</code>
کردن باید فراتر از درجه 
<code>3</code>
برویم.با استفاده از حلقه زیر در هر گام درجه 
<code>i</code>
ام را محاسبه و به ویژگی های مدل اضافه میکنیم تا اینکه 
<code>i</code>
برابر با
<code>10</code>
شود.
  </font>
</p>

```python
degree = 10
for deg in range(4 , degree+1):
    #calculate degree deg of X_poly_2d
    curr_feature = X_poly_2d ** deg
    # Concatenate curr_features with X_features
    X_features = np.concatenate([X_features , curr_feature] , axis = 1)
X_features
```
